## Rename columns to lowercase to resolve conflicts in query -- typical postgresql issue

```python
import psycopg2
from sqlalchemy import create_engine, inspect, text

# Replace with your actual connection details
DB_USER = 'postgres'
DB_PASS = 'postgres'
DB_HOST = 'localhost'
DB_PORT = '5432'
DB_NAME = 'vendordatabase'

# SQLAlchemy connection URL
engine = create_engine(f'postgresql+psycopg2://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}')
conn = engine.connect()
trans = conn.begin()

# Create inspector to get table/column metadata
inspector = inspect(engine)

try:
    for table_name in inspector.get_table_names():
        columns = inspector.get_columns(table_name)
        for col in columns:
            original = col['name']
            lower = original.lower()
            if original != lower:
                rename_stmt = text(f'ALTER TABLE {table_name} RENAME COLUMN "{original}" TO {lower};')
                print(f'Renaming {table_name}.{original} â†’ {lower}')
                conn.execute(rename_stmt)

    trans.commit()  # ğŸ”¥ Commit all changes
    print("âœ… All column renames committed.")
except Exception as e:
    trans.rollback()
    print(f'âŒ Error occurred: {e}')
finally:
    conn.close()
```
